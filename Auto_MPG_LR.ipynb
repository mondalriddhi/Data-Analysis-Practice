```python
x=file.drop(columns="mpg")
x
```

```python
y=file["mpg"]
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import  LinearRegression
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import r2_score
from sklearn.compose import make_column_transformer 
from sklearn.pipeline import make_pipeline
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
ohe=OneHotEncoder()
ohe.fit(x[["car name"]])
```

```python
#ohe.categories_
col_trans=make_column_transformer((OneHotEncoder(categories=ohe.categories_), ["car name"]), remainder='passthrough')
```

```python
lr=LinearRegression()
pipe=make_pipeline(col_trans, lr)
```

```python
pipe.fit(x_train, y_train)
```

```python
y_pred=pipe.predict(x_test)
```

```python
r2_score(y_test, y_pred)
```

```python
score=[]
for i in range(100):
    x_train, x_test, y_train, y_test=train_test_split(x, y, test_size=0.2, random_state=i)
    pipe.fit(x_train, y_train)
    ypred=pipe.predict(x_test)
    score.append(r2_score(y_test, ypred))
```

```python
import numpy as np
np.argmax(score)
score[36]
```

```python
x_train, x_test, y_train, y_test=train_test_split(x, y, test_size=0.2, random_state=np.argmax(score))
pipe.fit(x_train, y_train)
ypred=pipe.predict(x_test)
r2_score(y_test, ypred)
```

```python
import matplotlib.pyplot as plt
results=pd.DataFrame({"Actual": y_test, "Predicted": ypred})
import seaborn as sns
plt.figure(figsize=(8,6))
sns.scatterplot(x='Actual', y='Predicted', data=results)
plt.title('Actual VS Predicted Graph')
plt.xlabel('Actual values')
plt.ylabel('Predicted values')
plt.grid(True)
plt.show()
```
